```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

### Convolution : introduction

```python

# Load image using OpenCV
img = cv2.imread("datas/drone.png")
#img = cv2.imread("datas/wo.jpg")

img0 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Define a simple edge detection filter as a 3x3 matrix
filter = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])

# Apply the filter using OpenCV's filter2D function
filtered_img = cv2.filter2D(img, -1, filter)

# Plot the original and filtered images using Matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(img0, cmap='gray')
ax1.set_title('Original Image')
ax2.imshow(filtered_img, cmap='gray')
ax2.set_title('Filtered Image')
plt.show()
```

### Filtre moyenneur : lissage 

```python
#fm_img = cv2.imread("datas/ein.jpeg")
fm_img = cv2.imread("datas/brain.png")
fm0 = cv2.cvtColor(fm_img, cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(fm_img,-1,kernel)

plt.subplot(121),plt.imshow(fm0),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

### Filtre moyenneur avec Opencv

```python
# Apply a mean filter to the image using OpenCV's blur function
mean_img = cv2.blur(fm_img, (5, 5))

# Plot the original and mean-filtered images using Matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(cv2.cvtColor(fm_img, cv2.COLOR_BGR2RGB))
ax1.set_title('Original Image')
ax2.imshow(cv2.cvtColor(mean_img, cv2.COLOR_BGR2RGB))
ax2.set_title('Mean-Filtered Image')
plt.show()
```

### Filtre Median 

```python
# Load image using OpenCV
pois = cv2.imread('datas/brain.png')
#
# Apply a median filter to the image using OpenCV's medianBlur function
median_img = cv2.medianBlur(pois, 5)

# Plot the original and median-filtered images using Matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(cv2.cvtColor(pois, cv2.COLOR_BGR2RGB))
ax1.set_title('Original Image')
ax2.imshow(cv2.cvtColor(median_img, cv2.COLOR_BGR2RGB))
ax2.set_title('Median-Filtered Image')
plt.show()
```

### Floutage  : filtre gaussien

```python
#pois = cv2.imread('datas/ein.jpeg')
pois = cv2.imread("datas/wo.jpg")

#blurred_img = cv2.GaussianBlur(pois, (11, 11), 0)
blurred_img = cv2.GaussianBlur(pois, (9, 9), 0)

# Plot the original and blurred images using Matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(cv2.cvtColor(pois, cv2.COLOR_BGR2RGB))
ax1.set_title('Original Image')
ax2.imshow(cv2.cvtColor(blurred_img, cv2.COLOR_BGR2RGB))
ax2.set_title('Blurred Image')
plt.show()
```

### Affutage : sharpening :  v1

```python
img_s = cv2.imread("datas/wo.jpg")
kernel = np.array([[-1, -1, -1],
                   [-1,  9, -1],
                   [-1, -1, -1]])

# Apply the sharpen filter to the image using OpenCV's filter2D function
sharpened_img = cv2.filter2D(img_s, -1, kernel)

# Plot the original and sharpened images using Matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(cv2.cvtColor(img_s, cv2.COLOR_BGR2RGB))
ax1.set_title('Original Image')
ax2.imshow(cv2.cvtColor(sharpened_img, cv2.COLOR_BGR2RGB))
ax2.set_title('Sharpened Image')
plt.show()
```

### Afffutage V2

```python


# Load image using OpenCV
img_s = cv2.imread('datas/wo.jpg')

# Apply a Gaussian blur to the image
blur_img = cv2.GaussianBlur(img_s, (5, 5), 0)

# Subtract the blurred image from the original image to obtain the high-frequency components
high_freq_img = cv2.subtract(img_s, blur_img)

# Multiply the high-frequency components by a sharpening strength parameter
sharpened_img = cv2.addWeighted(img_s, 1.5, high_freq_img, -0.5, 0)

# Plot the original and sharpened images using Matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(cv2.cvtColor(img_s, cv2.COLOR_BGR2RGB))
ax1.set_title('Original Image')
ax2.imshow(cv2.cvtColor(sharpened_img, cv2.COLOR_BGR2RGB))
ax2.set_title('Sharpened Image')
plt.show()
```
