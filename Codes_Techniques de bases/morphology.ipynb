```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

### image


```python

img = cv2.imread("datas/plate.jpg")

img0 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img0)

plt.title('image')


plt.show()
```

### Erosion

```python
gray =  cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#define kernel

kernel = np.ones((5, 5), np.uint8)

kernel2 = np.ones((7, 7), np.uint8)


#perform tasks

erosion = cv2.erode(gray, kernel, iterations=3)


fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(img0, cmap='gray')
axs[0].set_title('Original Image')

axs[1].imshow(erosion, cmap='gray')
axs[1].set_title('Eroded Image')

plt.show()
```

### Dilatation 

```python
dilation = cv2.dilate(gray, kernel, iterations=3)


fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(img0, cmap='gray')
axs[0].set_title('Original Image')

axs[1].imshow(dilation, cmap='gray')
axs[1].set_title('dilated Image')

plt.show()
```

### ouverture 

```python
opening = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel)


fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(img0, cmap='gray')
axs[0].set_title('Original Image')

axs[1].imshow(opening, cmap='gray')
axs[1].set_title('Opened Image')

plt.show()
```

### fermeture 

```python
closing = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, kernel2)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(img0, cmap='gray')
axs[0].set_title('Original Image')

axs[1].imshow(closing, cmap='gray')
axs[1].set_title('Closed Image')

plt.show()

```
