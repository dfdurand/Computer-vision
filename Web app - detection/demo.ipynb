#### Face detection 

```python
import cv2
import matplotlib.pyplot as plt
# charger le fichier cascade  pour la dectection de faces 
face_cascade = cv2.CascadeClassifier('files/haarcascade_frontalface_default.xml')

# charger les images 
img = cv2.imread('datas/markle.jpg')


# Convertir en gris 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detecter les faces 
faces = face_cascade.detectMultiScale(gray, scaleFactor=2, minNeighbors=5)

# dessiner le rectangle autour de chaque face detecter 
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),5)

# afficher les faces sur l'images d'origines
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax1.set_title('Detected Faces')
ax2.imshow(cv2.cvtColor(cv2.imread('datas/markle.jpg'), cv2.COLOR_BGR2RGB))
ax2.set_title('Original Image')
plt.show()
```

#### smile detection

```python
import cv2
import matplotlib.pyplot as plt

# Load the Haar cascade file for smile detection
smile_cascade = cv2.CascadeClassifier('files/haarcascade_smile.xml')

# Load the images
img = cv2.imread('datas/sourires.jpg')
#img = cv2.imread('datas/smale.jpeg')
#img = cv2.imread('datas/sall.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect smiles in the grayscale image
smiles = smile_cascade.detectMultiScale(gray, scaleFactor=2.3, minNeighbors=10)

# Draw a rectangle around each detected smile
for (x,y,w,h) in smiles:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)

# Display the original and detected images side by side using matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax1.set_title('Detected Faces')
ax2.imshow(cv2.cvtColor(cv2.imread('datas/sourires.jpg'), cv2.COLOR_BGR2RGB))
ax2.set_title('Original Image')
plt.show()
```

#### eye detection

### face , eyes , smile detection 

```python
import cv2
import matplotlib.pyplot as plt

# Load the Haar cascade files for face, eye, and smile detection
face_cascade = cv2.CascadeClassifier('files/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('files/haarcascade_eye.xml')
smile_cascade = cv2.CascadeClassifier('files/haarcascade_smile.xml')

# Load the image
#img = cv2.imread('datas/gall.jpeg')
# img = cv2.imread('datas/b.png')
img = cv2.imread('datas/baby.jpg')


gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# dtection des faces 
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=7)

# Draw a rectangle around each detected face, and detect eyes and smiles within each face rectangle
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),4)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray, scaleFactor=1.1, minNeighbors=3)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),4)
    smiles = smile_cascade.detectMultiScale(roi_gray, scaleFactor=1.5, minNeighbors=22)
    for (sx,sy,sw,sh) in smiles:
        cv2.rectangle(roi_color,(sx,sy),(sx+sw,sy+sh),(0,0,255),4)

# Display the original and detected images side by side using matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax1.set_title('Detected Faces, Eyes, and Smiles')
ax2.imshow(cv2.cvtColor(cv2.imread('datas/baby.jpg'), cv2.COLOR_BGR2RGB))
ax2.set_title('Original Image')
plt.show()
```

### Upper and lower pose body detection 

```python
import cv2
import matplotlib.pyplot as plt

# Load the Haar cascade files for upper and lower body detection
upper_cascade = cv2.CascadeClassifier('files/haarcascade_upperbody.xml')
lower_cascade = cv2.CascadeClassifier('files/haarcascade_lowerbody.xml')

# Load the image
# img = cv2.imread('datas/gpose.jpg')
img = cv2.imread('datas/profilpose.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect upper body in the grayscale image
upper_bodies = upper_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=5)

# Draw a rectangle around each detected upper body
for (x,y,w,h) in upper_bodies:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)

# Detect lower body in the grayscale image
lower_bodies = lower_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=5)

# Draw a rectangle around each detected lower body
for (x,y,w,h) in lower_bodies:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)

# Display the original and detected images side by side using matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax1.set_title('Detected Bodies')
ax2.imshow(cv2.cvtColor(cv2.imread('datas/gpose.jpg'), cv2.COLOR_BGR2RGB))
ax2.set_title('Original Image')
plt.show()
```
